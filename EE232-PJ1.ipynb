```r
p <- c( 0.003, 0.004, 0.01, 0.05,0.1)
nodes <- 1000

g1 <- erdos.renyi.game(n=nodes,p=p[1],directed=F)
g2 <- erdos.renyi.game(n=nodes,p=p[2],directed=F)
g3 <- erdos.renyi.game(n=nodes,p=p[3],directed=F)
g4 <- erdos.renyi.game(n=nodes,p=p[4],directed=F)
g5 <- erdos.renyi.game(n=nodes,p=p[5],directed=F)


plot(degree.distribution(g1),main="Degree distribution of the network(p=0.003)",xlab="Degree",ylab="Frequency")
degreesVector <- degree(g1)
hist(degreesVector,main="Degree distribution of the network(p=0.003)",xlab="Degree",ylab="Frequency")
plot(degree.distribution(g2),main="Degree distribution of the network(p=0.004)",xlab="Degree",ylab="Frequency")
degreesVector <- degree(g2)
hist(degreesVector,main="Degree distribution of the network(p=0.004)",xlab="Degree",ylab="Frequency")
plot(degree.distribution(g3),main="Degree distribution of the network(p=0.01)",xlab="Degree",ylab="Frequency")
degreesVector <- degree(g3)
hist(degreesVector,main="Degree distribution of the network(p=0.01)",xlab="Degree",ylab="Frequency")
plot(degree.distribution(g4),main="Degree distribution of the network(p=0.05)",xlab="Degree",ylab="Frequency")
degreesVector <- degree(g4)
hist(degreesVector,main="Degree distribution of the network(p=0.05)",xlab="Degree",ylab="Frequency")
plot(degree.distribution(g5),main="Degree distribution of the network(p=0.1)",xlab="Degree",ylab="Frequency")
degreesVector <- degree(g5)
hist(degreesVector,main="Degree distribution of the network(p=0.1)",xlab="Degree",ylab="Frequency")

```

```r
cat("-----g1-----\n")
x1=mean(degree.distribution(g1))
y1=var(degree.distribution(g1))
cat("mean for g1 is",x1,"\n")
cat("variance for g1 is",y1,"\n")

cat("-----g2-----\n")
x2=mean(degree.distribution(g2))
y2=var(degree.distribution(g2))
cat("mean for g1 is",x2,"\n")
cat("variance for g1 is",y2,"\n")

cat("-----g3-----\n")
x3=mean(degree.distribution(g3))
y3=var(degree.distribution(g3))
cat("mean for g1 is",x3,"\n")
cat("variance for g1 is",y3,"\n")

cat("-----g4-----\n")
x4=mean(degree.distribution(g4))
y4=var(degree.distribution(g4))
cat("mean for g1 is",x4,"\n")
cat("variance for g1 is",y4,"\n")

cat("-----g5-----\n")
x5=mean(degree.distribution(g5))
y5=var(degree.distribution(g5))
cat("mean for g1 is",x5,"\n")
cat("variance for g1 is",y5,"\n")
```

```r
cat("g1 connectivity:",is.connected(g1),"\n")
cat("g2 connectivity:",is.connected(g2),"\n")
cat("g3 connectivity:",is.connected(g3),"\n")
cat("g4 connectivity:",is.connected(g4),"\n")
cat("g5 connectivity:",is.connected(g5),"\n")
```

```r
C1=C2=C3=C4=C5=0
for(i in 1:100){
  g1 = erdos.renyi.game(n=nodes,p=p[1],directed=FALSE)
  g2 = erdos.renyi.game(n=nodes,p=p[2],directed=FALSE)
  g3 = erdos.renyi.game(n=nodes,p=p[3],directed=FALSE)
  g4 = erdos.renyi.game(n=nodes,p=p[4],directed=FALSE)
  g5 = erdos.renyi.game(n=nodes,p=p[5],directed=FALSE)
  
  c1 = is.connected(g1)
  c2 = is.connected(g2)
  c3 = is.connected(g3)
  c4 = is.connected(g4)
  c5 = is.connected(g5)
    
  C1 = c(C1,c1)
  C2 = c(C2,c2)
  C3 = c(C3,c3)
  C4 = c(C4,c4)
  C5 = c(C5,c5)
 
}
C1_avg=mean(C1)
C2_avg=mean(C2)
C3_avg=mean(C3)
C4_avg=mean(C4)
C5_avg=mean(C5)

cat("Connectivity:",C1_avg,"\n")
cat("Connectivity:",C2_avg,"\n")
cat("Connectivity:",C3_avg,"\n")
cat("Connectivity:",C4_avg,"\n")
cat("Connectivity:",C5_avg,"\n")
```

```r
g1.components <- clusters(g1)
print(g1.components$csize)
# which is the largest component
ix <- which.max(g1.components$csize)
# get the subgraph correspondent to just the giant component
g1.giant <- induced.subgraph(g1, which(g1.components$membership == ix))
diameter(g1.giant)

g2.components <- clusters(g2)
print(g2.components$csize)
# which is the largest component
ix <- which.max(g2.components$csize)
# get the subgraph correspondent to just the giant component
g2.giant <- induced.subgraph(g2, which(g2.components$membership == ix))
diameter(g2.giant)
```

```r
n<-1000
p<-0
P=0
VALUE=0
for( i in 1:40 ){
    for(j in 1:100){
        g = erdos.renyi.game(n=n,p=p,directed=FALSE)
        g.components <- clusters(g)
        ix <- which.max(g.components$csize)
        value<-g.components$csize[ix]
        value<-value/1000
        VALUE = c(VALUE,value)
        P=c(P,p)
    }
    p<-p+0.0002
    
}
plot(P, VALUE, main="Normalized GCC size VS p", xlab="p ", ylab="Normalized GCC size",pch=1)
```

```r
c<- 0.5
P=N=VALUE=0
for(n in 100:1000){
    p<-c/n
    g = erdos.renyi.game(n=n,p=p,directed=FALSE)
    P=c(P,p)
    N=c(N,n)
    g.components <- clusters(g)
    value<-mean(g.components$csize)/n
    VALUE = c(VALUE,value)
    
}
plot(N,P, main="P VS n", xlab="nodes", ylab="p",pch=1)
plot(N,VALUE, main="Expected GCC size VS n", xlab="n", ylab="Expected GCC size",pch=1)
```
